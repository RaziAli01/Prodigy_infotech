```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('bmi.csv')
```

```python
print(df.head())
```

```python
plt.figure(figsize=(8, 6))
plt.hist(df['Age'], bins=10, color='skyblue', edgecolor='black')
plt.title('Age Distribution in a Population')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
```

```python
plt.show()
```
